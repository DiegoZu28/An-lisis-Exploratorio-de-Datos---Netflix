Pregunta 2.

En que año se publicaron mas peliculas y series?

Muestrelo con un grafico de barras

```python
netflix.dtypes[['Release_Date']]
```

```python
netflix['Date_N'] = pd.to_datetime(netflix['Release_Date'], format='mixed') # 'mixed' toma prueba varios formatos de fecha
```

```python
netflix.head()
```

```python
netflix.dtypes[['Date_N']]
```

```python
cantidad_años = netflix['Date_N'].dt.year.value_counts()
cantidad_años.sort_index(inplace = True)
cantidad_años
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x = cantidad_años.index, y = cantidad_años.values, hue = cantidad_años.index)
for i, v in enumerate(cantidad_años.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.title('Cantidad de películas y series por año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.show()
```
